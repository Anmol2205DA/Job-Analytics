Total number of jobs available as per location, Job_type, involvement, level,industry.

```python
col=['location','Job_type','involvement','level','industry']
for i in col:
    print(f"number of jobs available as per {i}")
    print(df.groupby(i)['jobid'].count())
    print()
```

```python
for i in col:
    if df[i].nunique()>7:
        plt.figure(figsize=(20,8))
    else:
        plt.figure(figsize=(8,6))
    sns.countplot(df[i],palette='Set1')
    plt.xticks(rotation=60)
    plt.show()
```

```python
df.head()
```

Total number of companies available as per location, job type, involvement, level, industry.

```python
col=['location','Job_type','involvement','level','industry']
for i in col:
    print(f"number of companies available as per {i}")
    result_df=df.groupby(i)['company'].count()
    print(result_df)
    if df[i].nunique()>7:
        plt.figure(figsize=(20,8))
    else:
        plt.figure(figsize=(8,6))
    sns.barplot(x=result_df.index,y=result_df.values,palette='Set1')
    plt.xticks(rotation=60)
    plt.show()
    print()
```

```python
df.head(1)
```

Total count of applicants for all jobs as per location , job_type, involvement, level, industry.

```python
col=['location','Job_type','involvement','level','industry']
for i in col:
    print(f"Total number of applicants as per {i}")
    result_df=df.groupby(i)['total_applicants'].sum()
    print(result_df)
    if df[i].nunique()>7:
        plt.figure(figsize=(20,8))
    else:
        plt.figure(figsize=(8,6))
    sns.barplot(x=result_df.index,y=result_df.values,palette='Set1')
    plt.xticks(rotation=60)
    plt.show()
    print()
```

```python
df.head(1)
```

Estimated sum of total number of employees as per location, job_type, involvement, level, industry.

```python
col=['location','Job_type','involvement','level','industry']
for i in col:
    print(f"Estimated number of employees as per {i}")
    result_df=df.groupby(i)['employees'].sum()
    print(result_df)
    if df[i].nunique()>7:
        plt.figure(figsize=(20,8))
    else:
        plt.figure(figsize=(8,6))
    sns.barplot(x=result_df.index,y=result_df.values,palette='Set1')
    plt.xticks(rotation=60)
    plt.show()
    print()
```

Estimated sum of number of followers of companies as per location, job_type,involvement,level,industry.

```python
col=['location','Job_type','involvement','level','industry']
for i in col:
    print(f"estimated number of followers as per {i}")
    result_df=df.groupby(i)['followers'].sum()
    print(result_df)
    if df[i].nunique()>7:
        plt.figure(figsize=(20,8))
    else:
        plt.figure(figsize=(8,6))
    sns.barplot(x=result_df.index,y=result_df.values,palette='Set1')
    plt.xticks(rotation=60)
    plt.show()
    print()
```
