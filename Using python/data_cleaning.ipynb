Removing all noise values and unwanted values.

```python
df.replace({' applicants':'',' followers':'',' employees':''},regex=True,inplace=True)
```

```python
df.replace({'Over ':'',' applicant':''},regex=True,inplace=True)
```

```python
df.employees.replace({',':'','Not given':'0'},regex=True,inplace=True)
```

```python
df.followers.replace({',':''},regex=True,inplace=True)
```

```python
df.head()
```

Total_applicants , employees , followers should be numeric columns.

```python
df.employees=df.employees.str.replace('+','',regex=True)
```

```python
df.employees=df.employees.str.split('-').str[0]
```

Changing datatypes to numeric.

```python
df['employees']=pd.to_numeric(df['employees'],errors='coerce')
```

```python
df['total_applicants']=pd.to_numeric(df['total_applicants'],errors='coerce')
```

```python
df['followers']=pd.to_numeric(df['followers'],errors='coerce')
```
