Total number of jobs available as per City, Job_Role, involvement, level, industry & Company size i.e., Employee_size.

```python
col=['City','Job_Role','Involvement','Level','Industry','Employee_Size']
for i in col:
    print(f"Number of jobs available as per {i}:")
    print(df.groupby(i)['ID'].count())
    print()
```

```python
for i in col:
    if df[i].nunique()>7:
        plt.figure(figsize=(20,8))
    else:
        plt.figure(figsize=(8,6))
    sns.countplot(df[i],palette='Set1')
    plt.xticks(rotation=60)
    plt.show()
```

Total number of companies available as per City, Job_role, Involvement, Level & Industry.

```python
col=['City','Job_Role','Involvement','Level','Industry']
for i in col:
    print(f"number of companies available as per {i}")
    result_df=df.groupby(i)['Company'].count()
    print(result_df)
    if df[i].nunique()>7:
        plt.figure(figsize=(20,8))
    else:
        plt.figure(figsize=(8,6))
    sns.barplot(x=result_df.index,y=result_df.values,palette='Set1')
    plt.xticks(rotation=60)
    plt.show()
    print()
```

Total count of applicants for all jobs as per City, job_role, involvement, level, industry.

```python
col=['City','Job_Role','Involvement','Level','Industry']
for i in col:
    print(f"Total number of applicants as per {i}")
    result_df=df.groupby(i)['Applicants'].sum()
    print(result_df)
    if df[i].nunique()>7:
        plt.figure(figsize=(20,8))
    else:
        plt.figure(figsize=(8,6))
    sns.barplot(x=result_df.index,y=result_df.values,palette='Set1')
    plt.xticks(rotation=60)
    plt.show()
    print()
```

Estimated sum of number of followers of companies as per City, job_role,involvement,level,industry.

```python
col=['City','Job_Role','Involvement','Level','Industry']
for i in col:
    print(f"estimated number of followers as per {i}")
    result_df=df.groupby(i)['Followers'].sum()
    print(result_df)
    if df[i].nunique()>7:
        plt.figure(figsize=(20,8))
    else:
        plt.figure(figsize=(8,6))
    sns.barplot(x=result_df.index,y=result_df.values,palette='Set1')
    plt.xticks(rotation=60)
    plt.show()
    print()
```

Detailed count of jobs grouped by a category further divided as per another category.

```python
col1=['Job_Role','Involvement','Level']
for i in col1:
    for j in col1:
        if i!=j:
           print(df.groupby(i)[j].value_counts())
           print('--------------------------------------------')
        else:
            pass
    print('-------------------------------------------------------------------------------')
```
